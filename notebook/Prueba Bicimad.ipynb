```python
from bicimad.bicimad import BiciMad

# 1. Crear una instancia de BiciMad con el mes, año, y opcionalmente una URL específica
bicimad_instance = BiciMad(month=2, year=23)  # Para datos de febrero de 2023
```

```python

# 2. Consulta de bicicletas desbloqueadas en una estación pero sin bloqueo registrado
cantidad_bicicletas_sin_bloqueo = bicimad_instance.count_unlocked_no_lock()
print(f"Cantidad de bicicletas desbloqueadas en una estación y no bloqueadas en ninguna: {cantidad_bicicletas_sin_bloqueo}")

# 3. Filtrar bicicletas de tipo de flota '1'
regular_fleet = bicimad_instance.filter_regular_fleet()
print("Bicicletas del tipo de flota '1':")
print(regular_fleet)

# 4. Calcular horas totales de uso por día
daily_hours_usage = bicimad_instance.day_time_usage()
print("Horas totales de uso de bicicletas por día:")
print(daily_hours_usage)

# 5. Calcular horas totales de uso por día de la semana
weekday_hours_usage = bicimad_instance.weekday_time_usage()
print("Horas totales de uso de bicicletas por día de la semana:")
print(weekday_hours_usage)

# 6. Obtener el número total de usos por día del mes
daily_usage_counts = bicimad_instance.total_usage_by_day()
print("Número total de usos de bicicletas por día del mes:")
print(daily_usage_counts)

# 7. Obtener el total de usos por fecha y estación de desbloqueo
usage_by_date_station = bicimad_instance.total_usage_by_date_and_station()
print("Total de usos de bicicletas por fecha y estación de desbloqueo:")
print(usage_by_date_station)

# 8. Obtener las estaciones de desbloqueo más populares
popular_stations = bicimad_instance.most_popular_stations()
print("Estaciones de desbloqueo más populares:", popular_stations)

# 9. Número total de usos de las estaciones más populares
total_usage_popular_stations = bicimad_instance.usage_from_most_popular_station()
print("Número total de usos de las estaciones más populares:", total_usage_popular_stations)

# 10. Resumen mensual de los datos
summary = bicimad_instance.resume()
print("Resumen mensual:\n", summary)
```
